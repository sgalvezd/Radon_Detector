```python
######### Import Libraries ###############
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from mpl_toolkits import mplot3d
import time
import scipy.integrate as sci
import math
from mpl_toolkits.mplot3d import Axes3D
import os
import imageio
from numpy import sin, cos, tan, arccos
from math import atan2, atan
import csv
import random as rd # 0<rd.random()<1
from matplotlib.patches import FancyArrowPatch
import pandas as pd
from collections import OrderedDict
import bisect
import datetime
import xlsxwriter
######### PLOT FUNCTIONS ###############

### PLOT CYLINDER MANTEL #################
def plot_mantel(R,H,px,pz,plane,jj):#R,H,nx,nz,(XY,XZ,ZY),window radius
    
    if plane == 'XY'or plane =='YX':
        e = 90
        az= 0
    elif plane == 'ZX' or plane == 'XZ':
        e = 0
        az= 90
    elif plane == 'YZ' or plane == 'ZY':
        e = 0
        az= 0
    else:
        e = 30
        az =30
    
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    
    #Cilinder
    z = np.linspace(-H/2,H/2,100)
    th = np.linspace(0,2*pi,100)
    Z, TH = np.meshgrid(z,th)
    X = R*np.cos(TH)
    Y = R*np.sin(TH)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.plot_surface(X,Y,Z,color='red',alpha=0.1)
    t = np.linspace(0,2*pi,100)
    ax.plot(R*np.sin(t),R*np.cos(t),t*0+H/2,'k--')
    ax.plot(R*np.sin(t),R*np.cos(t),t*0-H/2,'k--')
    #Plane
    x = np.linspace(-R,R,10)
    z = np.linspace(-H/2,H/2,10)
    X,Z = np.meshgrid(x,z)
    ax.plot_surface(X,X*0,Z,color='blue',alpha=0.3)
    ax.plot([R,-R,-R,R,R],[0,0,0,0,0],[-H/2,-H/2,H/2,H/2,-H/2],'k-')
    #points
    X,Z = np.meshgrid(px,pz)
    ax.scatter3D(X,X*0,Z,color='black',s=1)
    #views
    ax.view_init(elev=e, azim=az)
    #radiation
    xx = [x[6],0,z[8]]
    #xx = [0,0,0]
    Nth = 12
    Nphi = 12
    th  = np.linspace(0,pi,Nth)
    phi = np.linspace(0,2*pi,Nphi)
    for i in range(Nphi-1):
        for j in range(1,Nth-1):
            x=np.sin(th[j])*np.cos(phi[i])
            y=np.sin(th[j])*np.sin(phi[i])
            z=np.cos(th[j])
            #if jj==1:
            #    print("(%1.2f,%1.2f,%1.2f)"%(x,y,z))
            x=np.sin(th[j])*np.cos(phi[i])
            y=np.sin(th[j])*np.sin(phi[i])
            z=np.cos(th[j])
            tr,l = rectas_mantel(xx,th[j],phi[i],R,H)
            if tr==True:
                col = 'black'
                ax.plot3D([xx[0],l*x+xx[0]],[xx[1],l*y+xx[1]],[xx[2],l*z+xx[2]],color=col,marker='.',linewidth=0.5)
            else:
                col = 'red'
                ax.plot3D([xx[0],l*x+xx[0]],[xx[1],l*y+xx[1]],[xx[2],l*z+xx[2]],color=col,marker='^',linewidth=0.5)
    ax.set_xlim(-R,R)
    ax.set_ylim(-R,R)
    ax.set_zlim(-H/2,H/2)
    plt.show()
    return
#plot_mantel(R,H,px,pz,'ZY',1)
#plot_mantel(R,H,px,pz,'XY',2)
#plot_mantel(R,H,px,pz,'XZ',2)
#plot_mantel(R,H,px,pz,'',2)




### PLOT CYLINDER WINDOW #################
def plot_window(R,H,px,pz,plane,jj,r): #R,H,nx,nz,(XY,XZ,ZY),window radius
    
    if plane == 'XY'or plane =='YX':
        e = 90
        az= 0
    elif plane == 'ZX' or plane == 'XZ':
        e = 0
        az= 90
    elif plane == 'YZ' or plane == 'ZY':
        e = 0
        az= 0
    else:
        e = 30
        az =30
    
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    
    #Cilinder
    z = np.linspace(-H/2,H/2,100)
    th = np.linspace(0,2*pi,100)
    Z, TH = np.meshgrid(z,th)
    X = R*np.cos(TH)
    Y = R*np.sin(TH)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.plot_surface(X,Y,Z,color='red',alpha=0.1)
    t = np.linspace(0,2*pi,100)
    ax.plot(R*np.cos(t),R*np.sin(t),t*0+H/2,'k--')
    ax.plot(R*np.cos(t),R*np.sin(t),t*0-H/2,'k--')
    #window
    #X,Y = np.meshgrid(r*np.cos(t),r*np.sin(t))
    ax.plot(r*np.sin(t),r*np.cos(t),t*0-H/2,'g-',linewidth=5)
    #ax.plot_surface(X,Y,X*0-H/2,color='green',alpha=0.8)
    #Plane
    x = np.linspace(-R,R,10)
    z = np.linspace(-H/2,H/2,10)
    X,Z = np.meshgrid(x,z)
    ax.plot_surface(X,X*0,Z,color='blue',alpha=0.3)
    ax.plot([R,-R,-R,R,R],[0,0,0,0,0],[-H/2,-H/2,H/2,H/2,-H/2],'k-')
    #points
    X,Z = np.meshgrid(px,pz)
    ax.scatter3D(X,X*0,Z,color='purple',s=1)
    #views
    ax.view_init(elev=e, azim=az)
    #radiation
    xx = [0,0,0]
    Nth = 20
    Nphi = 20
    th  = np.linspace(0,pi,Nth)
    phi = np.linspace(0,2*pi,Nphi)
    for i in range(Nphi-1):
        for j in range(1,Nth-1):
            x=np.sin(th[j])*np.cos(phi[i])
            y=np.sin(th[j])*np.sin(phi[i])
            z=np.cos(th[j])
            #if jj==1:
            #    print("(%1.2f,%1.2f,%1.2f)"%(x,y,z))
            x=np.sin(th[j])*np.cos(phi[i])
            y=np.sin(th[j])*np.sin(phi[i])
            z=np.cos(th[j])
            tr,l = rectas_window(xx,th[j],phi[i],R,H,r)
            if tr==True:
                col = 'blue'
                ax.plot3D([xx[0],l*x+xx[0]],[xx[1],l*y+xx[1]],[xx[2],l*z+xx[2]],color=col,marker='*')
            else:
                col = 'red'
                ax.plot3D([xx[0],l*x+xx[0]],[xx[1],l*y+xx[1]],[xx[2],l*z+xx[2]],color=col,marker='*')
    ax.set_xlim(-R,R)
    ax.set_ylim(-R,R)
    ax.set_zlim(-H/2,H/2)
    plt.show()
    return
#plot_window(R,H,px,pz,'ZY',1,r)
#plot_window(R,H,px,pz,'XY',2,r)
#plot_window(R,H,px,pz,'XZ',2,r)
#plot_window(R,H,px,pz,'',2,r)

########### PLOT OF RANDOM RAYS IN RANDOM POINTS #################
def plot_random(R,H,N): #R,H,amount of events
    fig = plt.figure(figsize =(14, 9))
    ax = plt.axes(projection ='3d')
    z = np.linspace(-H/2,H/2,100)
    th = np.linspace(0,2*pi,100)
    Z, TH = np.meshgrid(z,th)
    X = R*np.cos(TH)
    Y = R*np.sin(TH)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.plot_surface(X,Y,Z,color='red',alpha=0.1)
    t = np.linspace(0,2*pi,100)
    ax.plot(R*np.sin(t),R*np.cos(t),t*0+H/2,'k--')
    ax.plot(R*np.sin(t),R*np.cos(t),t*0-H/2,'k--')

    x = np.zeros(N)
    y = x.copy()
    z = x.copy()
    for i in range(N):
        x[i],y[i],z[i] = point(R,H)
        a,b,c = direction('Cartesian')
        t = 50
        ax.quiver(x[i],y[i],z[i],t*a,t*b,t*c,color='red')

    ax.scatter3D(x,y,z,color='blue',marker='o',s=75)
    ax.set_xlim(-R,R)
    ax.set_ylim(-R,R)
    ax.set_zlim(-H/2,H/2)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.show()
    return

########## PLOT CONTOURS ######################
### N levels ##########
def plot_contour_N(X,Y,Z,N,colormap,color): #X,Y,Z,nlevels,cmap
    factor=0.5
    plt.figure(figsize =(factor*27.5*2/7, factor*86.5/7))
    contours = plt.contourf(X, Y, Z, N ,cmap=colormap)
    plt.clabel(contours, inline=True, fontsize=10)
    plt.colorbar()
    plt.grid()
    plt.xlabel('X')
    plt.ylabel('Z')
    plt.show()
    return
### explcit levels ####
def plot_contour_L(X,Y,Z,levels,colormap,mtitle,ctitle,save,name): 
    #X,Y,Z,explicit levels,cmap,main title,colorbar title,save?, name of the archive
    factor=0.6
    fig = plt.figure(figsize =(factor*27.5*2/7, factor*86.5/7))
    contourf = plt.contourf(X, Y, Z, levels ,cmap=colormap)
    contours = plt.contour(X, Y, Z, levels ,colors='black')
    plt.clabel(contours, inline=True, fontsize=15)
    plt.colorbar(contourf,label=ctitle)
    plt.title(mtitle,fontsize=15)
    #plt.ylim(0,70)
    plt.grid()
    plt.xlabel('X [mm]')
    plt.ylabel('Z [mm]')
    if save:
        fig.savefig(name+".pdf", bbox_inches='tight')
    plt.show()
    return
###### FUNCTIONS ##############

###### rays and only mantel ####################

def rectas_mantel(x,th,phi,R,H):#input a r(t)=x+(u,v,w)(phi,th)*t,t>=0 output True and Length or False
    if th==0 or th==pi:
        return False,H/2
    t = (np.sqrt(R**2-(x[0]*np.sin(phi)-x[1]*np.cos(phi))**2)-(x[0]*np.cos(phi)+x[1]*np.sin(phi)))/np.sin(th)
    if t<0:
        print(t)
        return 'ERROR t<0'
    if abs(x[2]+t*np.cos(th))<0.5*H:
        return True,t
    else:
        if 0<th<pi/2:
            return False,(H/2-x[2])/np.cos(th)
        elif pi/2<th<pi:
            return False,(-H/2-x[2])/np.cos(th)
        else:
            print(th,phi,t)
            return 'ERROR WTF'
        
###### rays and window ##########################

def rectas_window(x,th,phi,R,H,r):#input x=(x1,x2,x3), theta angle, phi angle, R,H, radius of window
    #a r(t)=x+(u,v,w)(phi,th)*t,t>=0 output True and Length or False
    if th==0 or th==pi:
        return True,H/2
    else:
        t = (np.sqrt(R**2-(x[0]*np.sin(phi)-x[1]*np.cos(phi))**2)-(x[0]*np.cos(phi)+x[1]*np.sin(phi)))/np.sin(th)
        if t<0:
            print("th=%1.2f"%(th/pi),"phi=%1.2f"%(0.5*phi/pi),"L=%1.2f"%t)
            print("point=(%1.5f,%1.5f,%1.5f)"%(x[0]/R,x[1]/R,2*x[2]/H))
            return "ERROR"
        elif abs(x[2]+t*np.cos(th))<0.5*H:
            #print("Manto")
            return True,t
        else:
            #print("notmanto")
            if 0<th<pi/2:
                #print("notmanto, arriba")
                return True,(H/2-x[2])/np.cos(th)
            elif pi/2<th<pi:
                #print("notmanto, abajo")
                t = (-H/2-x[2])/np.cos(th)
                x1 = x[0]+t*np.cos(phi)*np.sin(th)
                x2 = x[1]+t*np.sin(phi)*np.sin(th)
                #print(np.sqrt(x1**2+x2**2)<r)
                if np.sqrt(x1**2+x2**2)<r:
                    #print(np.sqrt(x1**2+x2**2))
                    return False,t
                else:
                    return True,t
            else:
                print("th=%1.2f"%th,"phi=%1.2f"%phi,"L=%1.2f"%t,
                      "point=(%1.2f,%1.2f,%1.2f)"%(x[0],x[1],x[2]))
                print('ERROR WTF')
                return
########### Efficiency for xi,zi ########################

def efficiency(x,z,R,H,Nth,Nphi,rank,r,w): #point x,point z,R,H,ntheta,nphi,rank,window radius,mantel/window
    #spherical mesh
    th  = np.linspace(0,pi,Nth)
    dth = th[1]-th[0]
    phi = np.linspace(0,2*pi,Nphi)
    dphi = phi[1]-phi[0]
    #mesh for plot

    cont = 0
    total = 0
    if w=='window':
            ##############################################################################
        for i in range(0,Nphi-1): #does not inclued th=2pi (repeated value th=0) ######## HERE ################
            ##############################################################################            
            for j in range(1,Nth-1): #does not include phi= 0,pi (do not touch the cilinder)
                point = [x,0,z] #point of plane's mesh
                tr,length = rectas_window(point,th[j],phi[i],R,H,r)
                total = total + 2*np.sin(th[j])*np.sin(0.5*dth)*dphi
                if tr and length<rank:
                    cont = cont+2*np.sin(th[j])*np.sin(0.5*dth)*dphi
    elif w=='mantel':
        for i in range(1,Nphi-1): #does not inclued th=2pi (repeated value th=0)
            for j in range(1,Nth-1): #does not include phi= 0,pi (do not touch the cilinder)
                point = [x,0,z] #point of plane's mesh
                tr,length = rectas_mantel(point,th[j],phi[i],R,H)
                total = total + 2*np.sin(th[j])*np.sin(0.5*dth)*dphi
                if tr and length<rank:
                    cont = cont+2*np.sin(th[j])*np.sin(0.5*dth)*dphi
    else:
        return "mantel or window"
    efficiency = cont/total
    return efficiency

####### FUNCTIONS FOR ALL POINTS IN THE PLANE #############
def Eff_plane(R,H,Nx,Nz,Nth,Nphi,r,rank,w): #R,H,nx,nz,nth,nphi,window radius,rank,window/mantel
    t0 = time.time()
    x = np.linspace(-R,R,Nx)
    z = np.linspace(-H/2,H/2,Nz)
    E = np.zeros((Nx-2,Nz-2))
    X = E.copy()
    Z = E.copy()
    for i in range(1,Nx-1):
        for j in range(1,Nz-1):
            X[i-1,j-1] = x[i]
            Z[i-1,j-1] = z[j]
            E[i-1,j-1] = efficiency(x[i],z[j],R,H,Nth,Nphi,rank,r,w)
            #print(E[i-1,j-1])
    t1 = time.time()
    timer(t0,t1)
    return X,Z,E

######### TIMER ############

def timer(t0,t1):
    dt = t1-t0
    h  = int(dt/3600)
    m  = int((dt-3600*h)/60)
    s  = dt-3600*h-60*m
    #t1 = time.time()
    print("simulation time: %d h %d m %1.2f s"%(h,m,s))
    return 

##### CSV WRITER ############################
def write_down(name,variable,X,Z,E):
    with open(name+'csv', 'w', encoding='UTF8', newline='') as f:

        writer = csv.writer(f)

        # write the header
        #writer.writerow(header)
        writer.writerow(["X","Z",variable])
        # write the data
        for i in range(len(X)):
            for j in range(len(Z)):
                writer.writerow([X[i,j],Z[i,j],E[i,j]])
    f.close()
    return
##### CSV READER ##############################
def read_csv(name):
    df = pd.read_csv(name)
    df = df.to_numpy()
    return df

#### ABSORPTION OF ENERGY ##############################
        
def absorption(d_max,rank):
    if d_max>=rank:
        return 0
    # Physical constantns
    pi = np.pi
    e = 1.60217662e-19 #[C], electron charge
    m_e = 9.10938356e-31 #[kg], electron mass
    r_e = 2.8179402894e-15 #[m], electron classical radius
    c = 299792458 #[m/s], speed of light in vaccum 
    N_a = 6.02214076e26 #[1/kmol]
    h = 6.62607015e-34 #[J/s]
    k = 2 * pi * N_a * r_e**2 * m_e * c**2/e
    print(k)

    #Properties of the absorving material (copper)
    Zeta = 0.49919 #for dry air Z/A=0.499, this
    A = 1
    Ip = 85.7
    rho = 1.2048 #[kg/m3]

    #properties of the incident particle (ALPHA PARTICLE)
    Mass = 6.64465723e-27 #[kg], mass of alpha particle
    ze = 2 # charge of incident particle in units of e, for alpha particle z=2
    E0 = 5.3 * 1e6
    #rank max distance traveled by the particle, app 40mm for alpha particles with 10MeV of E0
    #aEmean = 2.5e6
    s = m_e/Mass

    #absorption functions

    gamma = lambda beta: (1-beta**2)**(-0.5)  #divide by the charge of the electron e to convert from Joule to eV
    eta = lambda beta: beta* gamma(beta)
    W_max = lambda beta: (2*m_e*c**2/e) * (eta(beta))**2 / (1 + 2*s*(1+(eta(beta))**2)**0.5 + s**2)
    F = lambda beta: math.log((2*m_e*c**2/e)*(eta(beta))**2*W_max(beta)/(Ip**2))
    dbeta = lambda x, beta: (-k*rho*(Zeta/A)*ze**2/(Mass*c**2/e)) * eta(beta)**(-3)*(F(beta)-2*beta**2)

    x0=0
    beta0 = (1-(1+E0/(Mass*c**2/e))**(-2))**0.5
    max_step = 0.5e-3
    points = int(d_max/max_step)
    solution = sci.RK45(dbeta, x0, [beta0], d_max, max_step)
    for i in range(1,points):
        solution.step()
        if solution.status == 'finished':
            break
    return (gamma(solution.y[0])-1)*Mass*c**2/e  #solution.y[0] returns the final value of Energy E

########## ENERGY ABSORPTOIN IN Xi,Zi #####################################
def Energy(x,z,R,H,Nth,Nphi,rank,r,w):
    #spherical mesh
    th  = np.linspace(0,pi,Nth)
    dth = th[1]-th[0]
    phi = np.linspace(0,2*pi,Nphi)
    dphi = phi[1]-phi[0]
    cont = 0
    total = 0
    if w=='window':
        Ef = efficiency(x,z,R,H,Nth,Nphi,rank,r,w)
        for i in range(Nphi-1): #does not inclued th=2pi (repeated value th=0)
            for j in range(1,Nth-1): #does not include phi= 0,pi (do not touch the cilinder)
                point = [x,0,z] #point of plane's mesh
                tr,length = rectas_window(point,th[j],phi[i],R,H,r)
                #total = total + absorption(length)#2*np.sin(th[j])*np.sin(0.5*dth)*dphi
                #if tr and length<rank:
                if tr:
                    cont = cont + absorption(length/1000,rank)*2*np.sin(th[j])*np.sin(0.5*dth)*dphi
    elif w=='mantel':
        Ef = efficiency(x,z,R,H,Nth,Nphi,rank,r,w)
        for i in range(Nphi-1): #does not inclued th=2pi (repeated value th=0)
            for j in range(1,Nth-1): #does not include phi= 0,pi (do not touch the cilinder)
                point = [x,0,z] #point of plane's mesh
                tr,length = rectas_mantel(point,th[j],phi[i],R,H)
                #total = total + absorption(length)#2*np.sin(th[j])*np.sin(0.5*dth)*dphi
                #if tr and length<rank:
                if tr:
                    cont = cont + absorption(length/1000)*2*np.sin(th[j])*np.sin(0.5*dth)*dphi
    Energy = 1e-6*cont/Ef
    return Energy
############# ???????????????????????????????????? ##################################################
def Energy_plane(R,H,Nx,Nz,Nth,Nphi,r,rank,w): #R,H,nx,nz,nth,nphi,window radius,rank,window/mantel
    t0 = time.time()
    x = np.linspace(-R,R,Nx)
    z = np.linspace(-H/2,H/2,Nz)
    E = np.zeros((Nx-2,Nz-2))
    X = E.copy()
    Z = E.copy()
    for i in range(1,Nx-1):
        for j in range(1,Nz-1):
            X[i-1,j-1] = x[i]
            Z[i-1,j-1] = z[j]
            E[i-1,j-1] = Energy(x[i],z[j],R,H,Nth,Nphi,rank,r,w)
            #print(E[i-1,j-1])
    t1 = time.time()
    timer(t0,t1)
    return X,Z,E

##### DEFINE RANDOM POINT IN VOLUME ##########################

def point(R,H):
    r = R*rd.random()
    th = 2*pi*rd.random()
    z  = H*rd.random()-H/2
    return r*np.cos(th),r*np.sin(th),z

#### DEFINE RANDOM DIRECTION ####################################
def direction(k):
    if k=='Cartesian':
        th = pi*rd.random()
        phi = 2*pi*rd.random()
        x = np.sin(th)*np.cos(phi)
        y = np.sin(th)*np.sin(phi)
        z = np.cos(th)
        return x,y,z
    elif k=='Spherical':
        th = pi*rd.random()
        phi = 2*pi*rd.random()
        return th,phi
    else:
        return "WTFFF"
    
##### DICTIONARY #####################################################
def create_dict(name):
    db = read_csv(name)
    x = db[:,0]
    y = db[:,1]
    D = OrderedDict()
    for i in range(len(x)):
        D[x[i]] = y[i]
    return D
    
def get_value_dict(x,dictionary):
    D = dictionary
    minD = D[min(D.keys(), key = lambda key: abs(key-sk))],D[min(D.keys(), key = lambda key: abs(key-x))]
    #k = 
    del D[minD]
    minD2 = D[min(D.keys(), key = lambda key: abs(key-sk))],D[min(D.keys(), key = lambda key: abs(key-x))]
    m = (Dmin-Dmin2)/(x)
    return m*(x-x1)+minD
```
